## Modello di Markov a variabili latenti

Il modello latente di Markov (o Hidden Markov Model) assume la
dipendenza Markoviana sulle variabili latenti o non osservabili

La dipendenza osservata tra le risposte ad istanti temporali
successivi è spiegata da componenti non note e nel seguito indicate
con la notazione $U^{(t)}$ Unobserved per t=1...n

Il modello si utilizza anche quando le misurazioni disponibili (variabili
osservate) rappresentano in modo indiretto l’oggetto dello studio

Ad esempio nel caso di concetti latenti quali la soddisfazione, lo
stato di salute, il benessere, il grado di apprendimento, questi
vengono spesso misurati attraverso le risposte fornite a seguito di
una serie di interviste alle persone o test, ad esempio di
apprendimento, ripetuti nel tempo

In questo caso gli stati del processo latente identificano le
caratteristiche del fenomeno oggetto di studio

il modello si presta ad essere interpretato, nell’ambito delle tecniche
di machine learning, anche per classificare la popolazione di
riferimento in sottogruppi di unità che sono omogenee al loro interno
e disomogenee tra di loro.

Un’altra situazione si ha quando la variabile risposta è misurata con
errori e in questo caso la variabile latente è la reale variabile risposta,
e i veri stati del processo sono gli stati latenti e le transizioni tra gli
stati sono studiate tramite la matrice di transizione stimata per gli
stati latenti

In biologia, ad esempio, il modello di Markov a variabili latenti si
utilizza per l’analisi tramite marcatori dei siti degli istoni modificati
dove il modello viene impiegato per studiare la probabilità di
transizione tra le configurazioni osservate

Il modello precedente, come altri modelli con variabili latenti, si
compone di due parti:
- il modello di misura che coinvolge la parametrizzazione delle variabili
risposta condizionate alla corrispondente variabile non osservata
- il modello latente che coinvolge i parametri del processo latente
ovvero le probabilità iniziali e di transizione della catena di Markov.

I parametri del modello di misura sono le probabilità condizionate
della risposta al processo latente

$$\phi_{y|u}= f_{Y^{(t)|U^{(t)}} }(y|u)  $$
- t = 1..T
- u=1...k
- y= 0... c-1

Rispetto al modello di transizione illustrato nella sezione precedente
nella formulazione del modello di Markov a variabili latenti si assume
che queste ultime abbiano una distribuzione discreta non specificata.
In questo modo il modello è definito semiparametrico

In questo caso la variabile latente presenta k stati discreti
u, u = 1, . . . , k definiti stati latenti che rappresentano dei possibili
sotto gruppi da cui è formata la popolazione

I parametri del modello latente si denotano con la lettera u a pedice
in quanto riferiti alla componente non osservata e sono le probabilità
iniziali

$$\pi_u= f_{U^{(1)}}(u)$$
- u = 1...k

le probabilità di transizione che nel caso di catena omogenea si
denotano come
$$\pi_{u|\bar{u} }= f_{U^{(t)}|U^{(t-1)}} (u|\bar{u})$$
- t = 2...T
- $\bar{u}, u = 1...k$

Pertanto il numero dei parametri liberi del modello coinvolge le
seguenti quantità riferite al modello misura (c categorie della
variabile risposta) e al processo latente con k stati


La distribuzione di probabilità del processo latente $f_U(u)$ definisce la
probabilità a priori del processo di Markov e si determina

$$f_U (u)= \pi_{u^{(1)}} \pi_{u|\bar{u}}$$

La distribuzione delle variabili osservate $f_Y(y)$ in base alle assunzioni
precedenti si ricostruisce rispetto al modello di misura e al modello
latente nel modo seguente

$$f^Y(y)= \sum_u f_{Y|U}(y|u)f_U(u) $$
- $f_{Y|U}(y|u)$ - indica la distribuzione di probabilità condizionata Y
rispetto a U per ogni realizzazione y di Y .

Applicando la regola di Bayes si ottiene la distribuzione di
probabilità a posteriori per le variabili non osservate nel modo
seguente

$$f_{U|Y}(u|y)= \frac{f_{Y|U}(y|u)f_U(u)}{f_Y)(y)}$$

- In base alla probabilità a posteriori è possibile (come per i modelli
miscuglio per dati cross section) classificare ogni unità ad ogni
